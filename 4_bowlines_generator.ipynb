# Read files

Get list of dances in show order

```python
# puts the dance names into show order

dance_names = []
# Open the text file and read its content
with open(show_order_path, 'r') as file:
    for line in file:
        # Split each line into columns based on tabs
        columns = line.strip().split('\t')

        # Check if the line has at least two columns (index and dance name)
        if len(columns) >= 2:
            # Extract the dance name from the second column and append to the list
            dance_name = columns[1]
            dance_names.append(dance_name)

# Print the list of dance names
print(dance_names)
print(len(dance_names)) # should be 18
dance_names += ['Choreog', 'Exec'] # add choreog and exec
```

```python
# Read the master roster file for membership / drops
df = pd.read_csv(roster_path)
print(set(df.columns[7:-1]) == set(dance_names)) # rows should match up
df
```

# Generate Bow Lines

```python
# helper function. prints lists as a pretty string.
def print_list(List):
    string_list = string_names = str(List).replace("'", "").replace("[", "").replace("]", "")
    return string_list
```

```python
bow_lines = {dance: [] for dance in dance_names}
temp = []
for index, row in df.iterrows():
    # count the last one as the one we want

    for dance in reversed(dance_names):
        if row[dance] == 'x':
            #print("dancer", row['Email Address'], 'in', dance)
            kerb = row['Email Address'].split('@')[0]
            bow_lines[dance].append(kerb)
            break
```

```python
bow_lines_order = list(reversed(dance_names[:-2])) + dance_names[-2:]
```

Print result

```python
for line in bow_lines_order:
    names = bow_lines[line]
    string_names = print_list(names)
    print(line, '| ', len(names), ' | ', string_names)

#print(bow_lines)
```
